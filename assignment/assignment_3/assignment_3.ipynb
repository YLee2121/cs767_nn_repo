```python
# augment data 
# d1 original
d1 = pd.DataFrame(df2) 
n = len(d1)

# d2 x1
d2 = pd.DataFrame(d1)
new_df = sample_df_for_iris(n//3)
d2 = pd.concat([d2, new_df])
d2 = d2.reset_index(drop=True)

# d3 x2
d3 = pd.DataFrame(d2)
new_df = sample_df_for_iris(n//3)
d3 = pd.concat([d3, new_df])
d3 = d3.reset_index(drop=True)


# d3 x3 
d4 = pd.DataFrame(d3)
new_df = sample_df_for_iris(n//3)
d4 = pd.concat([d4, new_df])
d4 = d4.reset_index(drop=True)

# d5 x4
d5 = pd.DataFrame(d4)
new_df = sample_df_for_iris(n//3)
d5 = pd.concat([d5, new_df])
d5 = d5.reset_index(drop=True)
```

# Problem 4

```python
# XGBoost
start_time = time.time()
for df in [d1, d2, d3, d4, d5]:
    tmp = xgboost_accuracy(df)
    print(tmp)
print('time')
print(time.time() - start_time)
```

```python
# CATBoost
start_time = time.time()
for df in [d1, d2, d3, d4, d5]:
    tmp = catboost_accuracy(df)
    print(tmp)
print('time')
print(time.time() - start_time)
```

```python
# LightGBM
start_time = time.time()
for df in [d1, d2, d3, d4, d5]:
    tmp = lgbm_accuracy(df)
    print(tmp)
print('time')
print(time.time() - start_time)
```
