```python
import tensorflow.keras as KERAS

(train_data, train_label), (test_data, test_label) = KERAS.datasets.imdb.load_data(num_words=2000)
```

```python
print(train_data[0])
print(train_label[0])
max(max(sequence) for sequence in train_data)
```

```python
word_idx = KERAS.datasets.imdb.get_word_index()
reverse_word_idx = { val:key for (key, val) in word_idx.items()}
decode_review = ' '.join( [reverse_word_idx.get(i - 3, '?') for i in train_data[0]])
print(decode_review)
```

```python
import numpy as np

def vectorize_sequences(sequences, dimension=2000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] += 1
    return results

x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)
print(x_train[0])
print(x_train.shape)

y_train = np.asarray(train_label).astype('float32')
y_test = np.asarray(test_label).astype('float32')
print(y_train)
```

# Problem 1 Two Layers NN

```python
model = KERAS.models.Sequential()
model.add( 
    KERAS.layers.Dense(16, input_shape=(2000,))
)
model.add(
    KERAS.layers.Dense(16)
)
model.add(
    KERAS.layers.Dense(1, activation='sigmoid')
)

model.summary()
```

# Problem 2 Activation

```python
from src import model_activation
import time
```

```python
# relu activation
start_time = time.time()
m_relu = model_activation(KERAS.activations.relu)
m_relu.compile(
    optimizer=KERAS.optimizers.SGD(learning_rate=0.01),
    loss=KERAS.losses.BinaryCrossentropy(), 
    metrics=[KERAS.metrics.BinaryAccuracy()]
)
history = m_relu.fit(
    x_train,
    y_train,
    epochs=20,
    batch_size=256,
    validation_split=0.1
)
print(time.time() - start_time)
results = m_relu.evaluate(x_test, y_test)
print(results)
print(f'update times: {88 * 20}')
```

# Problem 4 Regularization

```python
# baseline NN
start_time = time.time()
model_base = KERAS.models.Sequential(
    [
        KERAS.layers.Dense(16, activation='relu', input_shape=(2000,)), 
        KERAS.layers.Dense(16, activation='relu'), 
        KERAS.layers.Dense(1, activation='sigmoid')

    ]

)

model_base.compile(
    optimizer=KERAS.optimizers.SGD(learning_rate=0.01),
    loss=KERAS.losses.BinaryCrossentropy(), 
    metrics=[KERAS.metrics.BinaryAccuracy()]
)

history = model_base.fit(
    x_train,
    y_train,
    epochs=20,
    batch_size=256,
    validation_split=0.1
)
print(time.time() - start_time)
results = model_base.evaluate(x_test, y_test)
print(results)
print(f'update times: {88 * 20}')
```

```python
# Dropout
start_time = time.time()
model_dropout = KERAS.models.Sequential(
    [
        KERAS.layers.Dense(16, activation='relu', input_shape=(2000,)), 
        KERAS.layers.Dropout(rate=0.75),
        KERAS.layers.Dense(16, activation='relu'), 
        KERAS.layers.Dense(1, activation='sigmoid')
    ]
)

model_dropout.compile(
    optimizer=KERAS.optimizers.SGD(learning_rate=0.01),
    loss=KERAS.losses.BinaryCrossentropy(), 
    metrics=[KERAS.metrics.BinaryAccuracy()]
)

history = model_dropout.fit(
    x_train,
    y_train,
    epochs=20,
    batch_size=256,
    validation_split=0.1
)
print(time.time() - start_time)
results = model_dropout.evaluate(x_test, y_test)
print(results)
print(time.time() - start_time)
print(f'update times: {88 * 20}')
```

```python
# Batch Norm
start_time = time.time()
model_batch_norm = KERAS.models.Sequential(
    [
        KERAS.layers.Dense(16, activation='relu', input_shape=(2000,)), 
        KERAS.layers.BatchNormalization(),
        KERAS.layers.Dense(16, activation='relu'), 
        KERAS.layers.Dense(1, activation='sigmoid')
    ]
)

model_batch_norm.compile(
    optimizer=KERAS.optimizers.SGD(learning_rate=0.01),
    loss=KERAS.losses.BinaryCrossentropy(), 
    metrics=[KERAS.metrics.BinaryAccuracy()]
)

history = model_batch_norm.fit(
    x_train,
    y_train,
    epochs=20,
    batch_size=256,
    validation_split=0.1
)
print(time.time() - start_time)
results = model_batch_norm.evaluate(x_test, y_test)
print(results)
print(f'update times: {88 * 20}')
```

```python
# Weight Initialization
start_time = time.time()
model_weight_init = KERAS.models.Sequential(
    [
        KERAS.layers.Dense(16, activation='relu', input_shape=(2000,)), 
        KERAS.layers.Dense(16, activation='relu', kernel_initializer=KERAS.initializers.RandomNormal(), bias_initializer=KERAS.initializers.RandomNormal()), 
        KERAS.layers.Dense(1, activation='sigmoid')
    ]
)

model_weight_init.compile(
    optimizer=KERAS.optimizers.SGD(learning_rate=0.01),
    loss=KERAS.losses.BinaryCrossentropy(), 
    metrics=[KERAS.metrics.BinaryAccuracy()]
)

history = model_weight_init.fit(
    x_train,
    y_train,
    epochs=20,
    batch_size=256,
    validation_split=0.1
)
print(time.time() - start_time)
results = model_weight_init.evaluate(x_test, y_test)
print(results)
print(f'update times: {88 * 20}')
```
